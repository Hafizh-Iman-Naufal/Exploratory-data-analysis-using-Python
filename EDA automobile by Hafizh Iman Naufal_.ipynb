## Kategori

```python

ds=pd.DataFrame(data['num-of-doors'].value_counts())
labels = list(ds.index)
sizes = ds
explode = (0, 0.1)  

fig1, ax1 = plt.subplots(figsize=(8,4))
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  

plt.show()
```

```python

ds=pd.DataFrame(data['horsepower-binned'].value_counts())
labels = list(ds.index)
sizes = ds
explode = (0, 0.1,0.5) 

fig1, ax1 = plt.subplots(figsize=(8,4))
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  

plt.show()
```
